```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2,2,100)

# la fonction
y = x ** 4 - x ** 2


# Parametre des axes et centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# tracer la fonction
plt.plot(x,y, 'y', label='y = x ** 4 - x ** 2')
plt.legend(loc='upper left')

# Montrer la fonction
plt.show()
```

```python
from scipy.optimize import minimize_scalar
def objective_function(x):

    return x ** 4 - x ** 2

res = minimize_scalar(objective_function)
```

```python
# On voit qu'il ne donne qu'une seule solution de valeur minimale de y, la positive
print(res)
```

```python
# Mais si on place des limites (bounds) entre -1 et 0 , alors il trouve la minimisation en < 0
res = minimize_scalar(objective_function, method='bounded', bounds=(-1, 0))
print(res)
```
