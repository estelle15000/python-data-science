<h1>Les 10 pays les plus gais </h1>

```python
df.groupby('Country or region')['Score'].max().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(12,6),color='yellow')

plt.title('Les 10 pays les plus gais')
```

<h1> Les 10 pays les plus tristes </h1>

```python
df.groupby('Country or region')['Score'].max().sort_values(ascending=False).tail(10).plot(kind='bar', figsize=(12,6),color='yellow')
plt.title('Les 10 pays les plus tristes')
```

<h1>Top 20 des pays avec le PIB par habitant le plus élevé</h1>

```python
df.groupby('Country or region')['GDP per capita'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='limegreen')
plt.title('Top 20 des pays avec le PIB par habitant le plus élevé')
```

<h1>20 pays avec le PIB par habitant le plus bas</h1>

```python
df.groupby('Country or region')['GDP per capita'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='limegreen')
plt.title('20 pays avec le PIB par habitant le plus bas')
```
