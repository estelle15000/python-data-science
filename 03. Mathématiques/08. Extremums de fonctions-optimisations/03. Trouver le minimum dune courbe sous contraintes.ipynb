# Trouver le minimum d'une courbe sous contrainte
# Optimisation

On utilise Scipy pour trouver le point minimum d'une fonction non lin√©aire.

Lien : https://caam37830.github.io/book/03_optimization/scipy_opt.html

```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
f = lambda x: x**2 + 1 # a x^T x + 0*x + 1

x = np.linspace(-2,2,500)
plt.plot(x, f(x))
plt.show()
```

```python
sol = opt.minimize_scalar(f)
sol
x = np.linspace(-2,2,500)
plt.plot(x, f(x))
plt.scatter([sol.x], [sol.fun], label='solution')
plt.legend()
plt.show()
```

Si vous souhaitez ajouter des contraintes telles que x>1 , vous pouvez utiliser des limites, tout en sp√©cifiant la m√©thode d√©limit√©e :

```python
sol = opt.minimize_scalar(f, bounds=(1, 1000), method='bounded')
x = np.linspace(-2,2,500)
plt.plot(x, f(x))
plt.scatter([sol.x], [sol.fun], label='solution')
plt.legend()
plt.show()
```

Pour les fonctions plus compliqu√©es, il existe plusieurs solutions d√©pendantes de l'intervalle choisie 

```python
f = lambda x : (x - 2)**2 * (x + 2)**2 - x

x = np.linspace(-3,3,500)
plt.plot(x, f(x))
plt.show()
```

```python
sol = opt.minimize_scalar(f)

plt.plot(x, f(x))
plt.scatter([sol.x], [sol.fun], label='solution')
plt.legend()
plt.show()
```
