## La dérivée

Rôle : Connaitre la pente d'une courbe en un point.

Intérêt : Déterminer une tendance en statistiques par exemple, linéariser...

```python
# On veut les dérivées successives de la fonction x²

import scipy.misc
import numpy as np

# On crée une fonction qui retourne y pour chaque x qu'on lui fourni
def f(x):
    return x*x # On reconnait bien là la fonction x²


# On crée un tableau qui va contenir les valeurs de X qu'on va "donner à manger" à la fonction f
X = np.array([-4,-3,-2,-1,0, 1, 2, 3, 4])

```

```python
# On affiche ici les valeurs de y pour x = -4, -3, -2 , -1, 0 , 1 , 2 , 3 , 4 pour la fonction x² 
print(f(X))
```

```python
# Maintenant on calcule les dérivées avec la fonction dédiée de scipy sur un intervalle positif 0, 1, 2 ,3 ,4
print(scipy.misc.derivative(f,np.array([0, 1, 2, 3, 4])))
# [0. 2. 4. 6. 8.]
```

```python
# On dessine la fonction x² pour voir à quoi elle ressemble

import matplotlib.pyplot as plt
x_cords = X
y_cords = f(X)


# On affiche la dérivée de la fonction x² sur une intervalle positive pour voir à quoi elle ressemble:
deriv_x = np.array([0, 1, 2, 3, 4])
deriv_y = np.array([0, 2, 4, 6, 8])


plt.plot(x_cords, y_cords)
plt.plot(deriv_x, deriv_y)
plt.show()

# On voit que la dérivée est une sorte de " résumé indicatif" de la pente de la fonction x²
```
