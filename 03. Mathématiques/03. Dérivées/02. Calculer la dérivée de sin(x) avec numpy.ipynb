## Calculer la dérivée de la fonction sin(x)

Grâce aux tableau habituel des dérivées, 
on sait que la dérivée de f = sin(x) 
est f'(x) = cos x pour toute valeur x.

Maintenant, on va apprendre à la calculer avec Numpy et à la tracer avec MatplotLib

Lien :
https://stackoverflow.com/questions/16841729/how-do-i-compute-the-derivative-of-an-array-in-python

```python
# On importe les librairies
import numpy as np
from matplotlib import pyplot as plt

# On crée une function sin(x) 
dx = np.pi/10
x = np.arange(0,2*np.pi,np.pi/10)

# On calcule la dérivée avec np.gradient et on la trace directement avec matPlotLib !
plt.plot(x,np.gradient(np.sin(x), dx), '-*', label='approx')

# we compare it with the exact first derivative, i.e. cos(x)
# On trace aussi la fonction cos(x) et on compare les deux pouvoir si c'est bon !
plt.plot(x,np.cos(x), label='exact')
plt.legend()
```
