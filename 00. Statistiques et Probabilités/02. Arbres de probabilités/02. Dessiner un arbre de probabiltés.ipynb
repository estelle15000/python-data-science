# Dessiner un arbre de probabilit√© 
avec python networkX par Estelle Derrien

```python
# Import des libs
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python


edges = [['A', 'B'], ['B', 'C'], ['B', 'D']]
G = nx.DiGraph() 
# G.add_node("K",pos=(0,0))
G.add_edges_from(edges)

position = nx.spring_layout(G)
plt.figure()
nx.draw(
    G,  edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()},pos=position
)
nx.draw_networkx_edge_labels(
    G, position,
    edge_labels={('A', 'B'): '1/3', 
                 ('B', 'C'): '1/4', 
                 ('B', 'D'): '1/4'},
    font_color='red'
)
plt.axis('off')
plt.show()

pos=nx.get_node_attributes(G,'pos')
print(pos)



```

On voit que l'arbre est correct mais pas bien positionn√© de gauche √† droite ( by Estelle)

```python


edges = [['A', 'B'], ['B', 'C'], ['B', 'D']]
G = nx.DiGraph() 
# G.add_node("K",pos=(0,0))
G.add_edges_from(edges)

position = nx.spring_layout(G)

plt.figure()

position["A"] = np.array([-3, 0])  # On sp√©cifie la position du node de cette fa√ßon
position["B"] = np.array([0, 0])  # On sp√©cifie la position du node de cette fa√ßon
position["C"] = np.array([3, 4])  # On sp√©cifie la position du node de cette fa√ßon
position["D"] = np.array([3, -4])  # On sp√©cifie la position du node de cette fa√ßon

nx.draw(
    G,  edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()},pos=position
)

nx.draw_networkx_edge_labels(
    G, position,
    edge_labels={('A', 'B'): '1/3', 
                 ('B', 'C'): '1/4', 
                 ('B', 'D'): '1/4'},
    font_color='red'
)
plt.axis('off')
plt.show()

pos=nx.get_node_attributes(G,'pos')
print(pos)
```
